## Slide 1

```python
df = pd.read_csv('B2S1.csv')
display(df.head())
xal3 = df['WL(nm)'].to_numpy()
yal3 = df['Abs'].to_numpy()

plt.figure(figsize=(10,7))
plt.plot(xal3, yal3, lw=0.5)
plt.xlabel('Wavelength (in nm)', size=12)
plt.ylabel('Absorption', size=12)
plt.title('Slide 1')
plt.grid()
plt.savefig('S1', dpi=300)
plt.show()
```

```python
xal, yal = xal3[250:800], yal3[250:800]
absavg = np.mean(yal)
plt.plot(xal, yal, lw=0.7)
plt.xlabel('Wavelength (in nm)', size=12)
plt.ylabel('Absorption', size=12)
plt.title('Slide 1 selected range')
plt.grid()
plt.savefig('S1p', dpi=300)
plt.show()

lams = xal*1e-9
rhoval, muval = 2.8e-8, 1.26e-6
delvals = np.sqrt(rhoval*lams/(np.pi*c*muval))
adiffs = yal
ts = 2.303*adiffs*delvals
ts1 = np.mean(ts)
tvar = np.var(ts)
sd1 = np.sqrt(tvar)
print(f'Average thickness: {ts1}, Error = {sd1}')
```
