# Atomic Force Microscopy (AFM)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
def linfit(x, m, c):
    return m*x + c
```

## 30 X 30

### Line 1

```python
df = pd.read_table('image30_1.txt')
x = df['X-Axis\''].to_numpy()
z = df['Z-Axis\''].to_numpy()
plt.title("30X30 Line 1")
plt.plot(x, z)
plt.xlabel("$x$ (nm)")
plt.ylabel("$z$ (nm)")
plt.grid()
plt.savefig('plot30_1.png', dpi=150)
plt.show()

x_ranges = np.array([0, 1550, 6740, 21510, 26500, 31000])
m = []
for i in range(len(x_ranges)-1):
    arr = np.where((x >= x_ranges[i]) & (x <= x_ranges[i+1]))[0]
    xi, zi = x[arr[0]:arr[-1]], z[arr[0]:arr[-1]]
    popt, pcov = curve_fit(linfit, xi, zi)
    m.append(popt[0])
m = np.array(m)
m_mean = m[2]
z_m = z - m_mean*x
plt.title("30X30 Line 1 modified")
plt.plot(x, z_m)
plt.xlabel("$x$ (nm)")
plt.ylabel("$z$ (nm)")
plt.grid()
plt.savefig('plot30_1m.png', dpi=150)
plt.show()
zm = []
for i in range(len(x_ranges)-1):
    arr = np.where((x >= x_ranges[i]) & (x <= x_ranges[i+1]))[0]
    xi, zi = x[arr[0]:arr[-1]], z_m[arr[0]:arr[-1]]
    zmi = np.mean(zi)
    zm.append(zmi)
zm = np.array(zm)
display('zm', zm)
zref = np.mean(np.array([zm[0], zm[2], zm[4]]))
print(f'Average surface height = {zref} nm')
ds, derr = [], []
arr = np.where((z_m<940))[0]
zd = z_m[arr]
zref = np.full_like(zd, zref)
d_val, d_err = np.mean(zref-zd), np.sqrt(np.var(zref-zd))
print(f'Average depth = {d_val} nm\nError: {d_err} nm')
```
