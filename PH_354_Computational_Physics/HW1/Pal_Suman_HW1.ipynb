## Exercise 2: Altitude of a satellite

```python
display(Image(filename='HW1Q2.png', width=900))
```

**SOLUTION:**

$\frac{1}{2} mv^2 = \frac{GMm}{r} \quad$ and $\quad v = \frac{2\pi r}{T}$

$$r^3 = \frac{GMT^2}{4\pi^2}$$

```python
def altitude_of_satellite(T):
    G = 6.67e-11
    M = 10**31
    R = 6400e3
    r = (G*M*T**2/(4*np.pi**2))**(1/3) - R
    return r

r1 = altitude_of_satellite(24*3600)*1e-3
print(f'Geosynchronous (24 hours): Altitude of satellite is {r1} km')
r2 = altitude_of_satellite(90*60)*1e-3
print(f'90 minutes: Altitude of satellite is {r2} km')
r3 = altitude_of_satellite(45*60)*1e-3
print(f'45 minutes: Altitude of satellite is {r3} km')
r4 = altitude_of_satellite(23.93*3600)*1e-3
print(f'Sidereal day (23.93 hours): Altitude of satellite is {r4} km')
print(f'Altitude difference asked in question 2.(c) is {r1-r4} km')
```

## Exercise 6: Catalan numbers

```python
display(Image(filename='HW1Q6.png', width=900))
```

**SOLUTION:**



```python
N = int(1e6)
def fnCatalan_numbers(n):
    c0 = 1
    if n > N:
        print('input number lesser equal to 1 billion')
    ni = 0
    cn = c0
    if n == 0:
        return c0
    else:
        while ni < n:
            cn1 = ((4*ni+2)/(ni+2))*cn
            cn = cn1
            ni = ni + 1
        return cn1

n1 = 5    # INPUT

print(f'for n={n1}, c{n1} = {fnCatalan_numbers(n1)}')

for i in range(101):
    ci = fnCatalan_numbers(i)
    print(f'n = {i}, c{i} = {ci}')
```

## Exercise 7: The Madelung constant

```python
display(Image(filename='HW1Q7_1.png', width=900))
display(Image(filename='HW1Q7_2.png', width=900))
```

**SOLUTION:**



```python
def Vijk(i, j, k):
    const = 1
    dijk = np.sqrt(i**2+j**2+k**2)
    return const/dijk

L = 50  # INPUT

Vtotal = 0
for i in range(-L, L+1):
    for j in range(-L, L+1):
        for k in range(-L, L+1):
            if abs(i)+abs(j)+abs(k) == 0:
                Velt = 0
            else:
                Velt = Vijk(i, j, k)
                if (i+j+k)%2 == 0:
                    Vtotal = Vtotal + Velt
                else:
                    Vtotal = Vtotal - Velt
                 
print(f'Value of Madelung constant is M = {Vtotal}')
```

## Exercise 8: The semi-empirical mass formula

```python
display(Image(filename='HW1Q8.png', width=900))
```

**SOLUTION:**



```python
def BndEn(Z, A):
    a1 = 15.67
    a2 = 17.23
    a3 = 0.75
    a4 = 93.2
    if A%2 == 0:
        if Z%2 == 0:
            a5 = 12
        else:
            a5 = -12
    else:
        a5 = 0
    return a1*A - a2*A**(2/3) - a3*Z**2/A**(1/3) - a4*(A-2*Z)**2/A + a5/(A)**(0.5)

Z1, A1 = 28, 58    # INPUT
print(f'For Z = {Z1} and A = {A1}, binding energy is {BndEn(Z1, A1)} MeV.')

def BEperA(Z, A):
    return BndEn(Z, A)/A
print(f'For Z = {Z1} and A = {A1}, binding energy per nucleon is is {BEperA(Z1, A1)} MeV.')
```

```python
def maxBE(Z):
    Aall = range(Z, 3*Z+1)
    BEall = []
    for A in Aall:
        BE_A = BEperA(Z, A)
        BEall.append(BE_A)
    for i in range(len(Aall)):
        if BEall[i] == max(BEall):
            return Aall[i]

Z2 = 92    # INPUT
print(f'For Z = {Z2}, most stable nucleus has A = {maxBE(Z2)}')

for Zi in range(1, 101):
    print(f'For Z = {Zi}, most stable nucleus has A = {maxBE(Zi)}')
```

## Exercise 9: Bionomial coefficients

```python
display(Image(filename='HW1Q9.png', width=900))
```

**SOLUTION:**



```python
def fn_bionomial(n, k):
    nk = 1
    for ni in range(1, n+1):
        nk = nk*ni
    for ki in range(1, k+1):
        nk = nk/ki
    for nki in range(1, n-k+1):
        nk = nk/nki
    return int(nk)

n1, k1 = 11, 0    # INPUT

print(f'bionomial({n1}, {k1}) = {fn_bionomial(n1, k1)}')
```

```python
for n in range(1, 21):
    nkstr = f''
    for k in range(n+1):
        nk = f' {fn_bionomial(n, k)} '
        nkstr = nkstr + nk
    print(nkstr)
```

```python
h60per100 = fn_bionomial(100, 60)/2**100
print(f'Probability of exact 60 heads out of 100 tosses is {h60per100*100} %.')

prob_b = 0
for i in range(60, 101):
    prob_i = fn_bionomial(100, i)/2**100
    prob_b = prob_b + prob_i
print(f'Probability of more than 60 heads out of 100 tosses is {prob_b*100} %.')
```

## Exercise 12: Deterministic chaos and the Feigenbaum plot

```python
display(Image(filename='HW1Q12_1.png', width=900))
display(Image(filename='HW1Q12_2.png', width=900))
display(Image(filename='HW1Q12_3.png', width=900))
```

**SOLUTION:**



```python
def fn_itr_map(r, x0, itr):
    x = x0
    for i in range(itr):
        xp = r*x*(1-x)
        x = xp
    return x
print(f'result for starting with x=0: {fn_itr_map(r=1, x0=0, itr=1000)}')

r_arr = np.arange(1, 4, 0.01)
x_arr = fn_itr_map(r=r_arr, x0=0.5, itr=1000)
plt.plot(r_arr, x_arr, 'o', ms=2)
plt.title('Feigenbaum plot')
plt.xlabel('r')
plt.ylabel('x')
plt.grid()
plt.show()
```

```python
itrs = np.arange(100, 1000, dtype=int)
x_arr = np.zeros(len(itrs))
for i in range(len(itrs)):
    x_arr[i] = fn_itr_map(r=2.5, x0=0.5, itr=itrs[i])
plt.plot(itrs, x_arr)
plt.title('Plot for r = 2.5')
plt.xlabel('iterations')
plt.ylabel('x')
plt.grid()
plt.show()
```

Thus, r = 2.5 is a fixed point.

```python
itrs = np.arange(1000, 1030, dtype=int)
x_arr = np.zeros(len(itrs))
for i in range(len(itrs)):
    x_arr[i] = fn_itr_map(r=3.2, x0=0.5, itr=itrs[i])
plt.plot(itrs, x_arr)
plt.title('Plot for r = 3.2')
plt.xlabel('iterations')
plt.ylabel('x')
plt.grid()
plt.show()
```

At r = 3.2, we get 2 periodic values (a limit cycle).

```python
itrs = np.arange(1000, 1030, dtype=int)
x_arr = np.zeros(len(itrs))
for i in range(len(itrs)):
    x_arr[i] = fn_itr_map(r=3.5, x0=0.5, itr=itrs[i])
plt.plot(itrs, x_arr)
plt.title('Plot for r = 3.5')
plt.xlabel('iterations')
plt.ylabel('x')
plt.grid()
plt.show()
```

At r = 3.5, we get 4 periodic values (a limit cycle).

```python
itrs = np.arange(1000, 1030, dtype=int)
x_arr = np.zeros(len(itrs))
for i in range(len(itrs)):
    x_arr[i] = fn_itr_map(r=3.8, x0=0.5, itr=itrs[i])
plt.plot(itrs, x_arr)
plt.title('Plot for r = 3.8')
plt.xlabel('iterations')
plt.ylabel('x')
plt.grid()
plt.show()
```

Here, there is no periodicity. It's a chaos.

**Conclusion:**

In the Feigenbaum plot,
- The points till r = 3, are fixed points.
- After discontinuity at r = 3, upto next discontinuity, the points (e.g. r = 3.2) are corresponding to limit cycles with 2 periodic values.
- The next part (e.g. r = 3.5) are corresponding to limit cylces with 4 periodic values.
- The final scattered part (e.g. r = 3.8) is corresponding to chaos.
