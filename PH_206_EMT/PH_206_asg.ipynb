```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Question 7

```python
def Phi_xyz(x, y, z, Nx, Ny):
    a = 1    # INPUT
    V = 1    # INPUT
    const = 8*V*a**2/np.pi**2
    phi_elt = 0
    for nx in range(1,Nx,2):
        X_x = np.sin(nx*np.pi*x/a)
        for ny in range(1,Ny,2):
            Y_y = np.sin(ny*np.pi*y/a)
            m = np.pi*np.sqrt(nx**2+ny**2)/a
            Z_z = (1-np.exp(-m*a))*np.exp(m*z) + (np.exp(m*a)-1)*np.exp(-m*z)
            phi_elt += (const/(nx*ny*np.sinh(m*a)))*X_x*Y_y*Z_z
    return phi_elt
```

**7. (b) Calculation of potential at center of cube:**

```python
nxny = [[3,3], [3,5], [5,5], [3,7], [5,7], [7,7], [3,9], [5,9], [7,9], [9,9]]
for ns in nxny:
    nxs = np.arange(1,ns[0],2)
    nys = np.arange(1,ns[1],2)
    terms_no = len(nxs) + len(nys)
    phi_nxny = Phi_xyz(0.5,0.5,0.5,ns[0],ns[1])
    print(f'nx={nxs}, ny={nys}, no. of terms = {terms_no}, Phi = {phi_nxny}')
```

```python
Nx1, Ny1 = 30, 30
L1, L2 = 0, 1
x1 = np.linspace(L1, L2, 50)
y1 = np.linspace(L1, L2, 50)
z1 = np.linspace(L1, L2, 50)
```

```python
x2, y2 = np.meshgrid(x1, y1)
z2 = 0.5        # INPUT
Phi2 = Phi_xyz(x2, y2, z2, Nx1, Ny1)
Phi_fixed = Phi_xyz(x2, y2, z2, Nx1, Ny1)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x2, y2, Phi2, cmap='viridis')
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel(f"$\Phi(x, y, z={z2})$")
ax.set_title(f"3D Surface Plot of $\Phi(x, y, z={z2})$")
plt.show()

y2, z2 = np.meshgrid(y1, z1)
x2 = 0.5        # INPUT
Phi2 = Phi_xyz(x2, y2, z2, Nx1, Ny1)
Phi_fixed = Phi_xyz(x2, y2, z2, Nx1, Ny1)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(y2, z2, Phi2, cmap='viridis')
ax.set_xlabel("$y$")
ax.set_ylabel("$z$")
ax.set_zlabel(f"$\Phi(x={x2}, y, z)$")
ax.set_title(f"3D Surface Plot of $\Phi(x={x2}, y, z)$")
plt.show()

x2, z2 = np.meshgrid(x1, z1)
y2 = 0.5        # INPUT
Phi2 = Phi_xyz(x2, y2, z2, Nx1, Ny1)
Phi_fixed = Phi_xyz(x2, y2, z2, Nx1, Ny1)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x2, z2, Phi2, cmap='viridis')
ax.set_xlabel("$x$")
ax.set_ylabel("$z$")
ax.set_zlabel(f"$\Phi(x, y={y2}, z)$")
ax.set_title(f"3D Surface Plot of $\Phi(x, y={y2}, z)$")
plt.show()
```
