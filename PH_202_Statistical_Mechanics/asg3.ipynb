```python
import numpy as np
import matplotlib.pyplot as plt
```

## Question 5

```python
z1 = np.arange(0.1, 2, 0.0015)
v0 = 1
M = int(1e3)
kB = 1
T = 10
P_z = lambda z: kB*T/(M*v0) * (M*np.log(1+z) + np.log(np.abs((1-z**M)/(1-z))))
lnz1, P1 = np.log(z1), P_z(z1)
plt.plot(lnz1, P1)
plt.axvline(0, color='black')
plt.title('Pressure vs ln(z) curve')
plt.xlabel('$\ln(z)$')
plt.ylabel('$P(z)$')
plt.grid()
plt.savefig('asg3plot1.png', dpi=150)
plt.show()

rho_z = lambda z: z/(M*v0) * (M/(1+z) + 1/(1-z) - M*z**(M-1)/(1-z**M))
rho1 = rho_z(z1)
plt.plot(lnz1, rho1)
plt.axvline(0, color='black')
plt.title('Number density vs ln(z) curve')
plt.xlabel('$\ln(z)$')
plt.ylabel('$\\rho(z)$')
plt.grid()
plt.savefig('asg3plot2.png', dpi=150)
plt.show()
```
