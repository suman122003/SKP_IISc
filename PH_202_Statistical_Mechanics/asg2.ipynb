```python
import numpy as np
import matplotlib.pyplot as plt
```

## Question 3

**Low temperature:**

```python
hw = 10
kB = 1
Z = lambda T: np.exp(hw/(24*kB*T))
T1 = np.linspace(1e-1,1,50)
plt.plot(T1, Z(T1))
plt.title('$T$ vs $Z(T)$ at low temperature')
plt.xlabel('$T$')
plt.ylabel('$Z(T)$')
plt.grid()
plt.show()
```

**High temperature:**

```python
hw = 10
kB = 1
Z = lambda T: np.sqrt(hw/(2*np.pi*kB*T)) * np.exp(np.pi**2*kB*T/(6*hw))
T1 = np.linspace(1e-1,16,100)
plt.plot(T1, Z(T1))
plt.title('$T$ vs $Z(T)$ at high temperature')
plt.xlabel('$T$')
plt.ylabel('$Z(T)$')
plt.grid()
plt.show()
```
