# RC Circuit as LPF and HPF

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df11 = pd.read_excel('expt_data.xlsx', sheet_name='RCLPF1')
df12 = pd.read_excel('expt_data.xlsx', sheet_name='RCLPF2')
df21 = pd.read_excel('expt_data.xlsx', sheet_name='RCHPF1')
df22 = pd.read_excel('expt_data.xlsx', sheet_name='RCHPF2')
```

```python
R, C = 10e3, 10e-9   # 10 kOhm, 10 nF
f0 = 1/(2*np.pi*R*C)
w0 = 1/(R*C)
```

## RC LPF

```python
def absTrclpf(w, w0):
    return w0/np.hypot(w0, w)
def argTrclpf(w, w0):
    return -np.arctan2(w, w0)
```

```python
f1 = np.linspace(0, 7000, 100)
w1 = 2*np.pi*f1
absT1 = absTrclpf(w1, w0)
plt.plot(f1, absT1, label='by theory')

f2 = df11['f'].to_numpy()
absT2 = df11['abs(T)'].to_numpy()

err = 0
for i in range(len(f2)):
    err += (absTrclpf(2*np.pi*f2[i], w0)-absT2[i])**2
err = np.sqrt(err/(len(f2)-1))
print(f'Error in expt values (std dev) = {err}.')

plt.plot(f2, absT2, 'o', label='expt data')
plt.title('RC LPF')
plt.xlabel('$f$ in Hz')
plt.ylabel('$|T(j\omega)|$')
plt.legend()
plt.grid()
plt.savefig('expt11.png', dpi=150)
plt.show()
```

```python
f1 = np.linspace(0, 7000, 100)
w1 = 2*np.pi*f1
argT1 = argTrclpf(w1, w0)
plt.plot(f1, argT1, label='by theory')

f2 = df12['f'].to_numpy()
argT2 = df12['arg(T)'].to_numpy()

err = 0
for i in range(len(f2)):
    err += (argTrclpf(2*np.pi*f2[i], w0)-argT2[i])**2
err = np.sqrt(err/(len(f2)-1))
print(f'Error in expt values (std dev) = {err}.')

plt.plot(f2, argT2, 'o', label='expt data')
y_value = -np.pi/2
plt.axhline(y=y_value, color='r', linestyle='--')
plt.text(1, y_value - 0.05, 'y = $-\pi/2$')
y_value = -np.pi/4
plt.axhline(y=y_value, color='r', linestyle='--')
plt.text(1, y_value - 0.05, 'y = $-\pi/4$')
plt.title('RC LPF')
plt.xlabel('$f$ in Hz')
plt.ylabel('arg$(T(j\omega))$')
plt.legend()
plt.grid()
plt.savefig('expt12.png', dpi=150)
plt.show()
```

## RC HPF

```python
def absTrchpf(w, w0):
    return w/np.hypot(w0, w)
def argTrchpf(w, w0):
    return np.arctan2(w0, w)
```

```python
f1 = np.linspace(0, 7000, 100)
w1 = 2*np.pi*f1
absT1 = absTrchpf(w1, w0)
plt.plot(f1, absT1, label='by theory')

f2 = df21['f'].to_numpy()
absT2 = df21['abs(T)'].to_numpy()

err = 0
for i in range(len(f2)):
    err += (absTrchpf(2*np.pi*f2[i], w0)-absT2[i])**2
err = np.sqrt(err/(len(f2)-1))
print(f'Error in expt values (std dev) = {err}.')

plt.plot(f2, absT2, 'o', label='expt data')
plt.title('RC HPF')
plt.xlabel('$f$ in Hz')
plt.ylabel('$|T(j\omega)|$')
plt.legend()
plt.grid()
plt.savefig('expt13.png', dpi=150)
plt.show()
```

```python
f1 = np.linspace(0, 7000, 100)
w1 = 2*np.pi*f1
argT1 = argTrchpf(w1, w0)
plt.plot(f1, argT1, label='by theory')

f2 = df22['f'].to_numpy()
argT2 = df22['arg(T)'].to_numpy()

err = 0
for i in range(len(f2)):
    err += (argTrchpf(2*np.pi*f2[i], w0)-argT2[i])**2
err = np.sqrt(err/(len(f2)-1))
print(f'Error in expt values (std dev) = {err}.')

plt.plot(f2, argT2, 'o', label='expt data')
y_value = np.pi/2
plt.axhline(y=y_value, color='r', linestyle='--')
plt.text(1, y_value - 0.05, 'y = $\pi/2$')
y_value = np.pi/4
plt.axhline(y=y_value, color='r', linestyle='--')
plt.text(1, y_value - 0.05, 'y = $\pi/4$')
plt.title('RC HPF')
plt.xlabel('$f$ in Hz')
plt.ylabel('arg$(T(j\omega))$')
plt.legend()
plt.grid()
plt.savefig('expt14.png', dpi=150)
plt.show()
```
