## Question 1

```python
gam1 = 6
T1 = 5
f, m, hcut = 1, 1, 1

def P1_wy(om, gam, T):
    I = (np.exp(1j*(om+gam)*T/2)*np.sin((om+gam)*T/2)/(om+gam) + 
                    np.exp(1j*(om-gam)*T/2)*np.sin((om-gam)*T/2)/(om-gam))
    return f**2/(2*m*om*hcut) * np.abs(I)**2

om1 = np.linspace(0.5, 15, 200)
plt.plot(om1, P1_wy(om1, gam1, T1))
plt.title(f'Plot for $\gamma$={gam1}')
plt.xlabel('$\omega$')
plt.ylabel('$P_1(\omega, \gamma, T)$')
plt.grid()
plt.show()
```

```python
gam1 = 6
T1 = 5
f, m, hcut = 1, 1, 1

def P1_wy(om, gam, T):
    I = ((np.sin((om+gam)*T/2)/(om+gam))**2 + 
            (np.sin((om-gam)*T/2)/(om-gam))**2 + 
            (np.cos(gam*T)-np.cos(om*T))*np.cos(gam*T)/(om**2-gam**2))
    return f**2/(2*m*om*hcut) * I

om1 = np.linspace(0.5, 15, 200)
plt.plot(om1, P1_wy(om1, gam1, T1))
plt.title(f'Plot for $\gamma$={gam1}')
plt.xlabel('$\omega$')
plt.ylabel('$P_1(\omega, \gamma, T)$')
plt.grid()
plt.show()
```

The probability peaks around $\gamma$ value.

Thus for getting transition, we need the $\gamma\rightarrow\omega$ limit.

## Question 5

```python
k0, r0 = 100, 10

def sig_k(k):
    expr1 = k*np.tan(k0*r0) - k0*np.tan(k*r0)
    expr2 = k0 + k*np.tan(k0*r0)*np.tan(k*r0)
    return (4*np.pi**2/k**2)*expr1**2/(expr2**2 + expr1**2)

k1 = np.linspace(2e-1, 5, 1000)
plt.plot(k1, sig_k(k1))
plt.axhline(0, color='black', lw=0.8)
plt.axvline(0, color='black', lw=0.8)
plt.title('Scattering cross section')
plt.xlabel('$k$')
plt.ylabel('$\sigma(k)$')
plt.grid()
plt.show()
```

The scattering cross section is higher at low energy (low $k$).

## Question 6

```python
import scipy.special as sps
j_l = lambda l, kr: sps.spherical_jn(l, kr)
y_l = lambda l, kr: sps.spherical_yn(l, kr)

kr1 = np.linspace(0.2, 10, 100)
for i in range(3):
    plt.plot(kr1, j_l(i, kr1), label=f'$l$={i}')
plt.title('Bessel function')
plt.xlabel('$kr$')
plt.ylabel('$j_l(kr)$')
plt.legend(loc='best')
plt.grid()
plt.show()

kr1 = np.linspace(2, 10, 100)
for i in range(3):
    plt.plot(kr1, y_l(i, kr1), label=f'$l$={i}')
plt.title('Neumann function')
plt.xlabel('$kr$')
plt.ylabel('$\eta_l(kr)$')
plt.legend(loc='best')
plt.grid()
plt.show()

kr1 = np.linspace(0.1, 15, 100)
kr2 = np.linspace(1, 15, 100)
plt.plot(kr1, j_l(1, kr1), label='Bessel function')
plt.plot(kr2, y_l(1, kr2), '--', label='Neumann function')
plt.title('Bessel and Neumann function')
plt.xlabel('$kr$')
plt.legend(loc='best')
plt.grid()
plt.show()
```

The spherical Neumann function diverges for $kr\rightarrow 0$.
